```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('Advertising.csv')

df.head()
df.shape
df.info
df.describe()
df.isnull().sum()
sns.pairplot(df,x_vars=["TV",'Radio','Newspaper'],y_vars='Sales',size=5,aspect=1,kind='scatter')
plt.show()
sns.regplot(x='TV',y='Sales',data=df)
sns.regplot(x='Newspaper',y='Sales',data=df)
sns.regplot(x='Radio',y='Sales',data=df)
plt.show()
sns.heatmap(df.corr(),cmap='YlGnBu',annot=True)
plt.show()
x=df['TV']
y=df['Sales']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,test_size=0.3,random_state=100)
x_train.head()
x_train.shape
y_train.head()
y_train.shape
import statsmodels.api as sm
x_train_sm=sm.add_constant(x_train)
lr=sm.OLS(y_train, x_train_sm).fit()
print(lr.params)
print(lr.summary())
plt.scatter(x_train,y_train)
plt.plot(x_train,6.948+0.054 *x_train,'r')
plt.show()
y_train_pred=lr.predict(x_train_sm)
res=(y_train-y_train_pred)
print(y_train,y_train_pred)
fig=plt.figure()
sns.distplot(res,bins=15)
fig.suptitle('Error terms',fontsize=15)
plt.xlabel('y_train-y_train_pred',fontsize=15)
plt.show()
plt.scatter(x_train,res)
plt.show()
x_test_sm=sm.add_constant(x_test)
y_test_pred=lr.predict(x_test_sm)
y_test_pred.head()
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
print('mean squared error',np.sqrt(mean_squared_error(y_test,y_test_pred)))
print('r score:',r2_score(y_test,y_test_pred))
print('mean absolute error:',mean_absolute_error(y_test,y_test_pred))
plt.scatter(x_test,y_test)
plt.plot(x_test,6.948+0.054*x_test,'r')
plt.show()
x=df['TV']
y=df['Sales']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,test_size=0.3,random_state=100)
import statsmodels.api as sm
x_train_sm=sm.add_constant(x_train)
lr=sm.OLS(y_train, x_train_sm).fit()
print(lr.params)
print(lr.summary())
plt.scatter(x_train,y_train)
plt.plot(x_train,12.165+0.124 *x_train,'r')
plt.show()
y_train_pred=lr.predict(x_train_sm)
res=(y_train-y_train_pred)
print(y_train,y_train_pred)
fig=plt.figure()
sns.distplot(res,bins=15)
fig.suptitle('Error terms',fontsize=15)
plt.xlabel('y_train-y_train_pred',fontsize=15)
plt.show()
plt.scatter(x_train,res)
plt.show()
x_test_sm=sm.add_constant(x_test)
y_test_pred=lr.predict(x_test_sm)
print('mean squared error',np.sqrt(mean_squared_error(y_test,y_test_pred)))
print('r score:',r2_score(y_test,y_test_pred))
print('mean absolute error:',mean_absolute_error(y_test,y_test_pred))
plt.scatter(x_test,y_test)
plt.plot(x_test,12.165+0.124*x_test,'r')
plt.show()
```
